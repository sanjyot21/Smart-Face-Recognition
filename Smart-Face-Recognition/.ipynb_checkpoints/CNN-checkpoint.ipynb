```python
from Model_architecture.modelArch import DenseArchs
import cv2
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
import numpy as np
import os
import pandas as pd

```

```python

n_classes = 2
arc = DenseArchs(n_classes)

face_model = arc.arch()
face_model.summary()
    
```

```python
x_data=[]
y_data = []
learning_rate = 0.01
epoch = 100
batch_size = 10        
people = sorted(os.listdir('people'))

for x in people:
    for i in os.listdir('people/'+x):
        img1 = cv2.imread('people'+'/'+x+'/'+i, -1)
        img = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
        
        img=cv2.resize(img,(16,16))
#         print(img.shape)
        img=img.astype('float')/255.0
#         print(img.size)
#         img=np.expand_dims(img,axis=0)
#         print(img.shape)
#         img = np.reshape(img,[128,128])
#         img[0,].reshape(img.shape[0:2]) 
        img=img.flatten()
        x_data.append(img)
        y_data.append(int(x[0:x.find('_')])-1)
#         print(x,i)


x_data=np.array(x_data,dtype='float')

y_data=np.array(y_data)
y_data=y_data.reshape(len(y_data),1)
X=x_data
y=to_categorical(y_data,num_classes=n_classes)
# print(X.shape)
face_model.compile(optimizer=Adam(learning_rate/epoch),loss='categorical_crossentropy', metrics=['accuracy'])
face_model.fit(X,y,batch_size=batch_size,validation_split=0.1,epochs=epoch,verbose=2)
face_model.save('Model/face_recog_model.h5')
print("Training Completed")
# print(y)
# face_model.compile(optimizer=Adam(learning_rate=learning_rate), loss='categorical_crossentropy', metrics=['accuracy'])
# face_model.fit(X, y=y, batch_size=batch_size, validation_split=0.1,epochs=epoch,verbose=2)
```

```python
from tabnanny import verbose
from unicodedata import decimal
import cv2
from FaceDetection.face_detection import face
from keras.models import load_model
import numpy as np
import os

import warnings
warnings.filterwarnings("ignore")
```

```python
label = None

person = None

def Create_labels():
    people = sorted(os.listdir('people'))
    students = {}
    attendance_count = {}
    for i in people:
        students[(int(i[0:i.find('_')])-1)]=i[i.find('_')+1:]
        attendance_count[(int(i[0:i.find('_')])-1)] = 0
    return students
```

```python
Create_labels()
people = sorted(os.listdir('peeps'))
model = load_model('Model/face_recog_model.h5')
print(people)
result=[]
peeps=[]
for x in people:
    for i in os.listdir('peeps/'+x):
        print(x,i)
        img1 = cv2.imread('peeps'+'/'+x+'/'+i, -1)
        img = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img, (224, 224))
        img = img.astype('float')/255.0
        img = np.expand_dims(img, axis=0)
        prediction = model.predict(img,verbose=2)
        peeps.append(i)
        result.append(int(np.argmax(prediction)))
        print(prediction[0][0]," ",prediction[0][1])
        print(int(np.argmax(prediction)))
        
```

```python
for yp,y in zip(result,peeps):
    print(y,end=" ")
    print(yp)
```
